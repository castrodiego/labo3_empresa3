```python
#Lea el Dataset base
data = "emp3_sellout_base_period_product.csv"
df = pd.read_csv(data)
df.head()
```

```python
# Como son 36 periodos que podemos usar para entrenar + validacion, use esa cantidad de lags teniendo en cuenta que LighGBM puede trabajar con valores nulos
# Aplico los lags + delta lags para product_id
num_lags = 36

# Lags para product_id
for i in range(1, num_lags + 1):
    df[f'product_id_lag_{i}'] = df.groupby('product_id')['tn'].shift(i)

# Delta lags para product_id
for i in range(1, num_lags):
    df[f'product_id_delta_lag_{i}'] = df[f'product_id_lag_{i}'] - df[f'product_id_lag_{i + 1}']



```

```python
# Aplico los lags + delta lags para cat1
num_lags = 36

# Lags para cat1
for i in range(1, num_lags + 1):
    df[f'cat1_lag_{i}'] = df.groupby('cat1')['tn'].shift(i)

# Delta lags para cat1
for i in range(1, num_lags):
    df[f'cat1_delta_lag_{i}'] = df[f'cat1_lag_{i}'] - df[f'cat1_lag_{i + 1}']


```

```python
# Aplico los lags + delta lags para cat2
num_lags = 36

# Lags para cat2
for i in range(1, num_lags + 1):
    df[f'cat2_lag_{i}'] = df.groupby('cat2')['tn'].shift(i)

# Delta lags para cat2
for i in range(1, num_lags):
    df[f'cat2_delta_lag_{i}'] = df[f'cat2_lag_{i}'] - df[f'cat2_lag_{i + 1}']
```

```python
# Aplico los lags + delta lags para cat3
num_lags = 36

# Lags para cat3
for i in range(1, num_lags + 1):
    df[f'cat3_lag_{i}'] = df.groupby('cat3')['tn'].shift(i)

# Delta lags para cat3
for i in range(1, num_lags):
    df[f'cat3_delta_lag_{i}'] = df[f'cat3_lag_{i}'] - df[f'cat3_lag_{i + 1}']
```
