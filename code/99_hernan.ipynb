```python
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#Lea el Dataset base
df = pd.read_csv("emp3_sellout_product_train.csv")
```

```python
# generamos nuevos tn y cust_request_tn con normalizacion

# Normalizacion MinMaxScaler lineal (0-1)
scaler = MinMaxScaler()
df[['tn_01', 'cust_request_tn_01']] = scaler.fit_transform(df[['tn', 'cust_request_tn']])

# Normalizacion standard con media y desvio
scaler = StandardScaler()
df[['tn_md', 'cust_request_tn_md']] = scaler.fit_transform(df[['tn', 'cust_request_tn']])
```

```python
#seleccionar que tipo de  tranformacion vamos a procesar

#//sin_norm=sin normalizacion //cero_uno=min max (0-1) //media_sd=standard media y desvio
tipos_norm = ['sin_norm', 'cero_uno', 'media_sd']

normalizacion = tipos_norm[0] 

if normalizacion == tipos_norm[0]:
    df = df.drop(['tn_01', 'tn_md','cust_request_tn_01','cust_request_tn_md'], axis=1)
    print('Normalizacion: '+normalizacion)
elif normalizacion == tipos_norm[1]:
    df = df.drop(['tn', 'tn_md','cust_request_tn','cust_request_tn_md'], axis=1)
    df = df.rename(columns={'tn_01': 'tn', 'cust_request_tn_01': 'cust_request_tn'})
    print('Normalizacion: '+normalizacion)
elif normalizacion == tipos_norm[2]:
    df = df.drop(['tn', 'tn_01','cust_request_tn','cust_request_tn_01'], axis=1)
    df = df.rename(columns={'tn_md': 'tn', 'cust_request_tn_md': 'cust_request_tn'})
    print('Normalizacion: '+normalizacion)

# Convertir 'periodo_fecha' a tipo datetime para asegurar el orden correcto
df['periodo_fecha'] = pd.to_datetime(df['periodo_fecha'])
# Ordenar por 'periodo_fecha' y 'product_id'
df_sorted = df.sort_values(by=['periodo_fecha', 'product_id'])
df.head()
```
